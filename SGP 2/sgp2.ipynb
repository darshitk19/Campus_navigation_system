```python
UMMlocation = (22.60027777777778, 72.819722222223)
map_UMM = folium.Map(location=UMMlocation, zoom_start=17, width="75%")
folium.TileLayer(
    tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
    attr='Tiles &copy; Esri'
).add_to(map_UMM)

# Display the map
map_UMM
```
