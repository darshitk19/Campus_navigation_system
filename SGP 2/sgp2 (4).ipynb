```python
class navigator:
    def __init__(self):
        self.geoResources = {}
        self.charusatlocation =(49.49226339787344, 8.487303256988524)
        self.position = 'w'
        self.destination = 'Haus17'

        for root, dirs, files in os.walk('HospitalNavigator/GeoResources'):
            for file in files:
                self.geoResources[file.split('.')[0]] = root+'/'+file

    def changeDestination(self,newDestination):
        self.destination = newDestination
        self.redrawMap()

    def changeStartPoint(self, newStartPoint):

        #self.position = newStartPoint #does not work yet
        print(f'Selected Start: {newStartPoint}; Selected Target: {self.destination}')
        #self.redrawMap()


    def drawPathWay(self,charusatmap):

      def switchPosition(coordinate):
        temp = coordinate[0]
        coordinate[0] = coordinate[1]
        coordinate[1] = temp
        return coordinate

      searchString = self.position + self.destination.split('Haus')[1]
      with open(self.geoResources[searchString]) as f:
           testWay = json.load(f)

      for feature in testWay['features']:
        path = feature['geometry']['coordinates']

      finalPath = list(map(switchPosition,path))
      folium.plugins.AntPath(finalPath).add_to(charusatmap)

    def drawBuilding(self,charusatmap):
      hauseOutline = self.geoResources[self.destination]
      folium.GeoJson(hauseOutline, name="geojson").add_to(charusatmap)

    def redrawMap(self):
        #print(f'position {self.position}, destination {self.destination}')
        charusatmap = folium.Map(location = self.charusatlocation, width = "75%", zoom_start = 17)
        self.drawPathWay(charusatmap)
        self.drawBuilding(charusatmap)
        display(charusatmap)

```

```python
myNavigator = Navigator()
```

```python
def displayWay(whereTo):
    myNavigator.changeDestination(whereTo)
```

```python
def changePosition(whereFrom):
    myNavigator.changeStartPoint(whereFrom)
```

```python
# Destination Selector
selectHouse_widget=ipywidgets.Select(

options=['charusatGround',
    'basketball_court',
    'rpcp',
    'cspit_cm',
    'cspit_ec',
    'depstar',
    'kkgirlhostel',
    'charusatLake',
    'cmpica',
    'iiim',
    'jcpgirlhostel',
    'mtin',
    'nidhisha',
    'pdpias',
    'charusathospital',
    'watertank',
    'workshop',
    'arip',
    'bdips',
    'centrallawn',
    ],
    value='charusatGround',
    description='Target',
    disabled=False)

# widget function
def selectHouse(way):
    if way == 'charusatGround' :
        displayWay('charusatGround' )
    if way == 'basketball_court':
        displayWay('basketball_court')
    if way == 'rpcp':
        displayWay('rpcp')
    if way == 'cspit_cm':
        displayWay('cspit_cm')
    if way == 'cspit_ec':
        displayWay('cspit_ec')
    if way == 'depstar':
        displayWay('depstar')
    if way == 'kkgirlhostel':
        displayWay('kkgirlhostel')
    if way == 'charusatLake':
        displayWay('charusatLake')
    if way == 'cmpica':
       displayWay('cmpica')
    if way == 'iiim':
        displayWay('iiim')
    if way == 'jcpgirlhostel' :
        displayWay('jcpgirlhostel' )
    if way == 'mtin':
        displayWay('mtin')
    if way == 'nidhisha':
        displayWay('nidhisha')
    if way == 'pdpias':
        displayWay('pdpias')
    if way == 'charusathospital':
        displayWay('charusathospital')
    if way == 'watertank':
        displayWay('watertank')
    if way == 'workshop':
        displayWay('workshop')
    if way == 'arip':
        displayWay('arip')
    if way == 'bdips':
        displayWay('bdips')
    if way == 'centrallawn' :
        displayWay('centrallawn' )

# Position Selector
selectPosition_widget=ipywidgets.Select(
    options=['Gate1_front', 'Gate2_back'],
    value='Gate2_back',
    description='Start',
    disabled=False)

def selectPosition(position):
    if position == 'Gate1_front':
        changePosition('f')
    if position == 'Gate2_back':
        changePosition('b')


#Initialization
ipywidgets.interact(selectPosition, position=selectPosition_widget)
ipywidgets.interact(selectHouse, way=selectHouse_widget)
```
